# Pendulo Doble

En general, un péndulo doble o doble péndulo es un sistema compuesto por dos péndulos, con el segundo colgando del extremo del primero. En el caso más simple, se trata de dos péndulos simples, con el inferior colgando de la masa pendular del superior. 

Normalmente se sobreentiende que nos referimos a un péndulo doble plano, con dos péndulos planos coplanarios. Este sistema físico posee dos grados de libertad y exhibe un rico comportamiento dinámico. Su movimiento está gobernado por dos ecuaciones diferenciales ordinarias acopladas. Por encima de cierta energía, su movimiento es caótico.

## Cinemática del Péndulo Doble

La cinemática se refiere a las relaciones de las partes del dispositivo, sin tener en cuenta las fuerzas. En cinemática solo se está tratando de encontrar expresiones para la posición, la velocidad y aceleración en términos de las variables que especifican el estado del dispositivo.

x = posición horizontal de la masa del péndulo

y = posición vertical de la masa del péndulo

$\theta$ = ángulo del péndulo (origen en la vertical hacia abajo, en sentido contrario a las agujas del reloj es positivo)

L = longitud de la varilla (constante)

Se coloca el origen en el punto del pivote del péndulo superior. Se considera que y aumenta hacia arriba. Se indica el péndulo superior por el subíndice 1, y el inferior por el subíndice 2. Se comienza usando trigonometría simple para escribir expresiones para las posiciones $x_1$, $y_1$, $x_2$, $y_2$ en términos de los ángulos $\theta_1$, $\theta_2$.

$$x_1 = L_1 sen \theta_1$$

$$y_1 = -L_1 cos \theta_1$$

$$x_2 = x_1 + L_2 sen \theta_2$$

$$y_2 = y_1 - L_2 cos \theta_2$$

La velocidad es la derivada con respecto al tiempo de la posición.

$$\dot{x}_1 = \dot{\theta}_1 L_1 cos \theta_1$$

$$\dot{y}_1 = \dot{\theta}_1 L_1 sen \theta_1$$

$$\dot{x}_2 = \dot{x}_1 + \dot{\theta}_2 L_2 cos \theta_2$$

$$\dot{y}_2 = \dot{y}_1 + \dot{\theta}_2 L_2 sen \theta_2$$

La aceleración es la segunda derivada.

\begin{equation} \tag{1}
\ddot{x}_1 = - \dot{\theta}_1^2 L_1 sen \theta_1 + \ddot{\theta}_1 L_1 cos \theta_1
\end{equation}

\begin{equation} \tag{2}
\ddot{y}_1 = \dot{\theta}_1^2 L_1 cos \theta_1 + \ddot{\theta}_1 L_1 sen \theta_1
\end{equation}

\begin{equation} \tag{3}
\ddot{x}_2 = \ddot{x}_1 - \dot{\theta}_2^2 L_2 sen \theta_2 + \ddot{\theta}_2 L_2 cos \theta_2
\end{equation}

\begin{equation} \tag{4}
\ddot{y}_2 = \ddot{y}_1 + \dot{\theta}_2^2 L_2 cos \theta_2 + \ddot{\theta}_2 L_2 sen \theta_2
\end{equation}

## Fuerzas en el doble péndulo

Se tratan a las dos masas del péndulo como partículas puntuales. Se definen las variables:

T = tensión en la varilla

m = masa del péndulo

g = constante gravitaional

Las fuerzas sobre la masa del péndulo superior son la tensión en la varilla superior $T_1$, la tensión en la varilla inferior $T_2$, y la gravedad $-m_1g$. Se escriben ecuaciones separadas para las fuerzas horizontales y verticales, ya que pueden ser tratadas independientemente. La fuerza neta sobre la masa es la suma de estas. Aquí se muestra la red de fuerza y se usa la ley de Newton $F = ma$.

\begin{equation} \tag{5}
m_1 \ddot{x}_1 = -T_1 sen \theta_1 + T_2 sen \theta_2
\end{equation}

\begin{equation} \tag{6}
m_1 \ddot{y}_1 = T_1 cos \theta_1 - T_2 cos \theta_2 cos \theta_2 - m_1 g
\end{equation}

Para el péndulo inferior, las fuerzas son la tensión en la varilla inferior $T_2$, y la gravedad $-m_2g$.

\begin{equation} \tag{7}
m_2 \ddot{x}_2 = - T_2 sen \theta_2
\end{equation}

\begin{equation} \tag{8}
m_2 \ddot{y}_2 = T_2 cos \theta_2 - m_2g
\end{equation}

## Método directo para encontrar ecuaciones de movimiento

Ahora se hacen algunas manipulaciones algebraicas con el objetivo de encontrar expresiones para $\ddot{\theta}_1$, $\ddot{\theta}_2$ en términos de $\theta_1$, $\dot{\theta}_1$, $\theta_2$, $\dot{\theta}_2$. Se comienza resolviendo las ecuaciones (7), (8) por $T_2 sen\theta_2$ y $T_2 cos \theta_2$ y luego sustituyéndolo por las ecuaciones (5) y (6).

\begin{equation} \tag{9}
m_1 \ddot{x}_1 = - T_1 sen \theta_1 - m_2 \ddot{x}_2
\end{equation}

\begin{equation} \tag{10}
m_1 \ddot{y}_1 = T_1 cos \theta_1 - m_2 \ddot{y}_2 - m_2g - m_1g
\end{equation}

Al multiplicar la ecuación (9) por $cos \theta_1$ y la ecuación (10) por $sen \theta_1$ y reorganizar se obtiene

\begin{equation} \tag{11}
T_1 sen \theta_1 cos \theta_1 = - cos \theta_1 (m_1 \ddot{x}_1 + m_2 \ddot{x}_2)
\end{equation}

\begin{equation} \tag{12}
T_1 sen \theta_1 cos \theta_1 = sen \theta_1 (m_1 \ddot{y}_1 + m_2 \ddot{y}_2 + m_2g + m_1g)
\end{equation}

Esto nos lleva a la ecuación

\begin{equation} \tag{13}
sen \theta_1 (m_1 \ddot{y}_1 + m_2 \ddot{y}_2 + m_2 g + m_1 g) = - cos \theta_1 (m_1 \ddot{x}_1 + m_2 \ddot{x}_2
\end{equation}

A continuación, se multiplica la ecuación (7) por $cos \theta_2$ y la ecuación (8) por $sen \theta_2$ y reorganizar para obtener

\begin{equation} \tag{14}
T_2 sen \theta_2 cos \theta_2 = - cos \theta_2 (m_2 \ddot{x}_2)
\end{equation}

\begin{equation} \tag{15}
T_2 sen \theta_2 cos \theta_2 = sen \theta_2 (m_2 \ddot{y}_2 + m_2 g)
\end{equation}

lo que lleva a

\begin{equation} \tag{16}
sen \theta_2 (m_2 \ddot{y}_2 + m_2 g) = - cos \theta_2 (m_2 \ddot{x}_2)
\end{equation}

A continuación, se necesita resolver las ecuaciones (13) y (16) para $\ddot{\theta}_1$ y $\ddot{\theta}_2$ en términos de $\theta_1$, $\dot{\theta}_1$, $\theta_2$ y $\dot{\theta}_2$. Se debe tener en cuenta también las definiciones dadas por las ecuaciones (1-4), de modo que se tienen 2 ecuaciones (13, 16) y 2 ecuaciones ($\ddot{\theta}_1$, $\ddot{\theta}_2$). Las ecuaciones de movimiento para el péndulo doble son

$$\ddot{\theta}_1 = \frac{- g (2 m_1 + m_2) sen \theta_1 - m_2 g sen (\theta_1 - 2 \theta_2) - 2 sen (\theta_1 - \theta_2) m_2 (\dot{\theta}_2^2 L_2 + \dot{\theta}_1^2 L_1 cos (\theta_1 - \theta_2))}{L_1 (2 m_1 + m_2 - m_2 cos (2 \theta_1 - 2 \theta_2))}$$

$$\ddot{\theta}_2 = \frac{2 sen (\theta_1 - \theta_2) (\dot{\theta}_1^2 L_1 (m_1 + m_2) + g (m_1 + m_2) cos \theta_1 + \dot{\theta}_2^2 L_2 m_2 cos (\theta_1 - \theta_2))}{L_2 (2 m_1 + m_2 - m_2 cos (2 \theta_1 - 2 \theta_2))}$$

## Solución numérica

Las ecuaciones anteriores se acercan ahora a la forma necesaria para el método de Runge Kutta. El paso final es convertir estas dos ecuaciones de 2° orden en cuatro ecuaciones de 1° orden. Se define el parámetro de primeras derivadas como variables separadas:

$\omega_1$ = velocidad angular de la varilla superior

$\omega_2$ = velocidad angular de la varilla inferior

Entonces se puede escribir las cuatro ecuaciones de 1er orden:

$$\dot{\theta}_1 = \omega_1$$

$$\dot{\theta}_2 = \omega_2$$

$$\dot{\omega}_1 = \frac{- g (2 m_1 + m_2) sen \theta_1 - m_2 g sen (\theta_1 - 2 \theta_2) - 2 sen (\theta_1 - \theta_2) m_2 (\omega_2^2 L_2 + \omega_1^2 L_1 cos (\theta_1 - \theta_2))}{L_1 (2 m_1 + m_2 - m_2 cos (2 \theta_1 - 2 \theta_2))}$$

$$\dot{\omega}_2 = \frac{2 sen (\theta_1 - \theta_2) (\omega_1^2 L_1 (m_1 + m_2) + g (m_1 + m_2) cos \theta_1 + \omega_2^2 L_2 m_2 cos (\theta_1 - \theta_2))}{L_2 (2 m_1 + m_2 - m_2 cos (2 \theta_1 - 2 \theta_2))}$$

Estas ecuaciones son necesarias para aplicar el método de Runge-Kutta para la solución numérica del sistema.

El objetivo será resolver estas ecuaciones de movimiento del péndulo simple utilizando el método de Runge-Kutta 4 para analizar su comportamiento al darle ciertos valores iniciales. En los siguientes cálculos se tomarán $w_1 = 0$ y $w_2 = 0$, lo que significa que no se le está aplicando velocidad en cuanto comienza a funcionar el sistema.

Primero se observará el comportamiento del péndulo con ángulos pequeños, siendo estos $\theta_1 = 0.5$ y $\theta_2 = 1$.

```python
import numpy as np
import matplotlib.pyplot as plt

def derivs(state, t, L1, L2, m1, m2, g):
    theta1, omega1, theta2, omega2 = state
    delta = theta1 - theta2

    den1 = L1 * (2 * m1 + m2 - m2 * np.cos(2 * delta))
    den2 = (L2 / L1) * den1

    dtheta1_dt = omega1
    dtheta2_dt = omega2
    
    domega1_dt = ((- g * (2 * m1 + m2) * np.sin(theta1) - 
                   m2 * g * np.sin(theta1 - 2 * theta2) - 
                   2 * np.sin(delta) * m2 * (omega2**2 * L2 + omega1**2 * L1 * np.cos(delta))) / den1)
    
    domega2_dt = ((2 * np.sin(delta) * (omega1**2 * L1 * (m1 + m2) + g * (m1 + m2) * np.cos(theta1) + omega2**2 * L2 * m2 * np.cos(delta))) / den2)

    return np.array([dtheta1_dt, domega1_dt, dtheta2_dt, domega2_dt])

def rk4_step(state, t, dt, derivs, L1, L2, m1, m2, g):
    k1 = dt * derivs(state, t, L1, L2, m1, m2, g)
    k2 = dt * derivs(state + 0.5 * k1, t + 0.5 * dt, L1, L2, m1, m2, g)
    k3 = dt * derivs(state + 0.5 * k2, t + 0.5 * dt, L1, L2, m1, m2, g)
    k4 = dt * derivs(state + k3, t + dt, L1, L2, m1, m2, g)
    return state + (k1 + 2*k2 + 2*k3 + k4) / 6

# Parámetros
L1 = 1.0
L2 = 1.0
m1 = 1.0
m2 = 1.0
g = 9.81
dt = 0.01
t_max = 20.0

# Condiciones iniciales
theta1_0 = 0.5
omega1_0 = 0.0
theta2_0 = 1
omega2_0 = 0.0
state = np.array([theta1_0, omega1_0, theta2_0, omega2_0])

# Tiempo
t = np.arange(0, t_max, dt)

# Almacenamiento de resultados
states = np.zeros((len(t), 4))
states[0] = state

# Integración numérica usando RK4
for i in range(1, len(t)):
    state = rk4_step(state, t[i-1], dt, derivs, L1, L2, m1, m2, g)
    states[i] = state

plt.plot(t, states[:, 0], label='theta 1')
plt.plot(t, states[:, 2], label='theta 2')
plt.xlabel('Tiempo (t)')
plt.ylabel('Ángulo (radianes)')
plt.legend()
plt.grid(True)
plt.show()
```

Para ángulos pequeños, un péndulo se comporta como un sistema lineal. Cuando los ángulos son pequeños en el péndulo doble, el sistema se comporta como el resorte doble lineal. En el siguiente gráfico, se pueden ver curvas de lissajous, las cuales son una familia de curvas descritas por senos y cosenos. Estas tienen diversas aplicaciones en física, astronomía y otras ciencias.

```python
plt.plot(states[:, 0], states[:, 2])
plt.xlabel('Ángulo theta 1 (radianes)')
plt.ylabel('Ángulo theta 2 (radianes)')
plt.grid(True)
plt.show()
```

Para ángulos grandes, el péndulo no es lineal y el sistema se vuelve caótico. Un sistema se define caótico cuando este evoluciona en el tiempo y que es muy sensible a las condiciones iniciales. Esto significa que al camniar un poco las condiciones iniciales, los resultados serán completamente diferentes. Para ver esto, se observará que sucede con cambiar el ángulo del péndulo inferior en 0.002 radianes, es decir, $\theta_2 = 1.998$, $\theta_2 = 2$ y $\theta_2 = 2.002$.
